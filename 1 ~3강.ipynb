###도미와 빙어 합치기

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

df3 = pd.DataFrame({'length':length, 'weight':weight})
df3.head()
df3.info()
df3.shape
df3.describe()
```

```python
plt.scatter(length,weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
fish_data = [[l, w ] for l, w in zip(length, weight)]
```

```python
fish_data
```

정답준비(지도학습)


```python
fish_target = [1]*35 + [0]*14
```

```python
fish_target

# 1의 갯수를 세는 함수
print(fish_target.count(1))
print(fish_target.count(0))

print(fish_target)
```

K-최근접 이웃

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
print(f'{kn.score(fish_data, fish_target)*100}%')
```

### 새로운 생선 예측
  - array([1]) = bream
  - array([0]) = smelt

```python
kn.predict([[30, 600]])
```

### 무조건 bream

최근접이웃의 갯수를 KNeighborsClassifier() 함수에서 입력 파라미터 조건을 'n_neighbors= 갯수'으로 설정하면 k-최근접 이웃의 계산시 이웃의 갯수를 조정할 수 있다. 입력을 하지 않으면 5개로 설정된다.

```python
# 계산시 필요한 갯수를 전체 샘플 갯수로 판단했기 때문에 전부 bream(True , 1)로 판단한다.
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
fish_length = bream_length + smelt_length
fish_weight = bream_weight + smelt_weight

df3 = pd.DataFrame({'length':fish_length, 'weight': fish_weight})
df3.head()
df3.info()
df3.shape
df3.describe()
```

```python
# row_stack 예시
fish_data2 = np.row_stack([fish_length, fish_weight])
fish_data2
```

```python
fish_data = np.column_stack([fish_length, fish_weight])
fish_data
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
fish_target
```

```python
np.ones((2,3))
```

```python
np.full((2,3),9)
```

###사이킷런으로 데이터 나누기

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target,stratify=fish_target, random_state=42) ## stratify 파라미터는 타겟값을 입력함으로써 랜덤하게 데이터를 분할 할때 데이터의 비중을 균질하게 만들어준다
```

### 수상한 도미

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
print(f'정확도 : {kn.score(test_input, test_target)*100} %')
```

```python
print(f'예측값 : {kn.predict([[25,150]])} (1 = bream , 0 = smelt)')
```

```python
distances, indexes = kn.kneighbors([[25,150]])  ## kn.kneighbors() 메소드를 사용하여 거리와 인덱스를 추출해 낼 수 있다.
print(f'거리 : {distances}')
print(f'인덱스 : {indexes}')
```

```python
print(train_input[indexes,0])
print(train_input[indexes,1])
```

```python
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')

plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlim((0,1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 표준화하기

```python
mean = np.mean(train_input, axis=0)  ## axis = 0 행을 따라 평균 전체를 계산 (열들의 전체평균)  axis =1 열을 따라 전체 평균을 계산(행들의 전체평균)
std = np.std(train_input, axis=0)

print(mean, std)
```

```python
print(train_input.shape)
print(train_input)
```

### 넘파이의 브로드캐스팅

```python
train_scaled = (train_input - mean) / std
train_scaled
```

```python
train_scaled.shape
```

### 테스트 세트 혹은 예측하고 싶은 데이터의 표준화
훈련 세트의 데이터의 평균과 표준편차로 표준화한다.

```python
new = ([25, 150] - mean) / std

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(new)
```

### 전처리 데이터에서 모델 훈련

```python
kn.fit(train_scaled, train_target)

test_scaled = (test_input - mean) / std
print(f'정확도 : {kn.score(test_scaled, test_target)*100} %')

print(f'예측값 : {kn.predict([new])} (1 = bream , 0 = smelt)')

distances, indexes = kn.kneighbors([new])  ## kn.kneighbors() 메소드를 사용하여 거리와 인덱스를 추출해 낼 수 있다.
print(f'거리 : {distances}')
print(f'인덱스 : {indexes}')

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
